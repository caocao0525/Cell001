```python
# convert all the data
import glob
import matplotlib
from matplotlib import image
from matplotlib.image import imsave

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
import imageio
from PIL import Image, ImageOps
import cv2

from skimage import color
from skimage import io
from skimage.color import gray2rgb, rgb2gray
```

```python
images = glob.glob('../img_before_conv/*.jpg')
```

```python
### par1 images ###

# image_num = 71  #late1
# image_num = 231  #sparse1
# image_num = 21  #sparse2
# image_num = 1  #sparse3
# image_num = 51  #early1
# image_num = 31  #late2
# image_num = 1  #early2
image_num = 1  #late3

for image in images:
    with open(image, "rb") as file:
        img = Image.open(file)
        gray_img = ImageOps.grayscale(img)
#         save_name = '../img_converted/par1_extra/par1_early_'+'{:03}'.format(image_num)+'.jpg'
#         save_name = '../img_converted/par1_extra/par1_sparse_'+'{:03}'.format(image_num)+'.jpg'
        save_name = '../img_converted/par1_extra/par1_late_'+'{:03}'.format(image_num)+'.jpg'
        gray_img.save(save_name)
        image_num+=1
        
print("All data were converted.")
```

```python
### wt images ###

# image_num = 1  #sparse1
# image_num = 51  #early1
# image_num = 31  #late1
# image_num = 1  #early2
image_num = 1  #late2

for image in images:
    with open(image, "rb") as file:
        img = Image.open(file)
        gray_img = ImageOps.grayscale(img)
#         save_name = '../img_converted/wt_extra/wt_early_'+'{:03}'.format(image_num)+'.jpg'
#         save_name = '../img_converted/wt_extra/wt_sparse_'+'{:03}'.format(image_num)+'.jpg'
        save_name = '../img_converted/wt_extra/wt_late_'+'{:03}'.format(image_num)+'.jpg'
        gray_img.save(save_name)
        image_num+=1
        
print("All data were converted.")
```
