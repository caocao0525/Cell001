```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3],[0.2989, 0.5870, 0.1140])
# based on the formula Y' = 0.2989 R + 0.5870 G + 0.1140 B
img = mpimg.imread('../data/par1/par1_001.jpg')
gray = rgb2gray(img)
plt.imshow(gray)
plt.show()
```
